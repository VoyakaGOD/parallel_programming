#### Уравнение:

$$y'' = a(y^3-y), \;\; y(-10) = y(10) = \sqrt{2}$$

#### Схема Нумерова:

$$F_m = y_{m+1} - 2y_m + y_{m+1} - \mu[f(y_{m+1}) + 10f(y_m) + f(y_{m-1})] = 0$$

$$f(y) = y^3 - y, \;\; \mu = \frac{ah^2}{12}$$

#### Производные:

$A_m = \frac{\partial F_m}{\partial y_{m-1}} = 1 - \mu(3y_{m-1}^2-1)$

$B_m = \frac{\partial F_m}{\partial y_m} = -2 - 10\mu(3y_m^2-1)$

$C_m = \frac{\partial F_m}{\partial y_{m+1}} = 1 - \mu(3y_{m+1}^2-1)$

#### Метод Ньютона:

$$J_F(\vec{y}^{(k)})\delta = -F(\vec{y}^{(k)})$$

$$\vec{y}^{(k+1)} = \vec{y}^{(k)} + \delta$$

На каждой итерации метода Ньютона получаем СЛАУ с 3-х диагональной матрицей:

$A_m \delta_{m-1} + B_m \delta_m + C_m \delta_{m+1} = -F_m$

Которую решаем через циклическую редукцию.

```python
import matplotlib.pyplot as plt

with open("y.csv") as file:
    y = [float(value) for value in file.read().split()]

x = [-10 + t/len(y) for t in range(len(y))]

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("y(x), a = 10, h = 0.01")
plt.show()
```

Врема ряботы для $a=10^{10}, h = 3 \cdot 10^{-6}$.

```python
T1 = 5.90177
ps = [1,       2,       3,       4,      5,       6,      7,       8]
Tp = [8.54387, 5.42207, 3.84369, 3.2081, 2.59693, 2.1899, 2.04497, 1.95483]

plt.hlines(T1, ps[0], ps[-1], [1,0.55,0], label="$T_1$")
plt.plot(ps, Tp, label="$T_p$")
plt.xlabel("p")
plt.ylabel("T")
plt.title(r"$a=10^{10}, h = 3 \cdot 10^{-6}$")
plt.legend()
plt.show()

S = [T1 / t for t in Tp]
plt.plot(ps, S)
plt.xlabel("p")
plt.ylabel("S")
plt.title("Speedup")
plt.show()

E = [s/(p+1) for p,s in enumerate(S)]
plt.plot(ps, E)
plt.xlabel("p")
plt.ylabel("S")
plt.title("Effeciency")
plt.show()
```

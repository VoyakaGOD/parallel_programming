## Найдём время работы на 100.000.000 итераций

```python
import matplotlib.pyplot as plt

ps  = [1,        2,        3,        4,        8]
nfs = [0.197239, 0.205882, 0.22698,  0.252619, 0.280197]
fsh = [0.200594, 0.379725, 0.567179, 0.820819, 1.68357]

plt.plot(ps, nfs, label="no false sharing")
plt.plot(ps, fsh, label="false sharing")
plt.xlabel("threads")
plt.ylabel("time, s")
plt.legend()
plt.show()
```

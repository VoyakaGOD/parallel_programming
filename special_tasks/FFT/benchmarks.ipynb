Найдём оптимальный threshold для FFT на 4-х потоках

```python
import matplotlib.pyplot as plt

threshold = [8, 16, 32, 64, 128, 256]
T = [0.281199, 0.257976, 0.244952, 0.222546, 0.207892, 0.26753]

plt.plot(threshold, T)
plt.xlabel("time, s")
plt.ylabel("threshold")
```

Измерим производительность на $2^{18}$ отсчётах

```python
T1 = 0.291111
ps = [1,        2,        4,        8]
Tp = [0.409866, 0.229444, 0.218531, 0.190511]

plt.plot(ps, [T1/t for t in Tp])
plt.xlabel("p")
plt.ylabel("S")
plt.title("Speedup")
plt.show()

plt.plot(ps, [T1/t/p for p, t in zip(ps, Tp)])
plt.xlabel("p")
plt.ylabel("E")
plt.title("Effeciency")
plt.show()
```

## Генератор отсчётов для FFT

```python
import matplotlib.pyplot as plt
import numpy as np

# params
T = 1
N = 2**18
t = np.arange(0, T, T / N)

#signal
meander = np.sign(np.sin(2 * np.pi * 5 * t))
sine = np.sin(2 * np.pi * 20 * t)
signal = meander * sine
step = max(1, N // 1024)

plt.figure(figsize=(10,4))
plt.plot(t[::step], signal[::step])
plt.title("Сигнал f(t)")
plt.xlabel("Время, с")
plt.ylabel("Амплитуда")
plt.title("Сигнал")
plt.show()

freq = np.fft.fftfreq(N, d=T)
spectrum = np.fft.fft(signal)
spec_amp = np.abs(spectrum) / N

plt.figure(figsize=(10,4))
plt.plot(freq[::step], spec_amp[::step])
plt.title("Амплитудный спектр")
plt.xlabel("Частота, Гц")
plt.ylabel("Амплитуда")
plt.title("Спектр")
plt.show()

np.savetxt("signal.csv", signal, delimiter=',')
```

## Спектр после внешнего fft

```python
ext_spectrum = np.loadtxt("spectrum.csv", delimiter=',')

plt.figure(figsize=(10,4))
plt.plot(freq[::step], ext_spectrum[::step])
plt.title("Амплитудный спектр")
plt.xlabel("Частота, Гц")
plt.ylabel("Амплитуда")
plt.title("Спектр")
plt.show()
```
